#####5. Gunakan for loop dan fungsi range() untuk mencetak angka genap dari 2 hingga 20. Pastikan Anda menggunakan parameter ketiga pada range(start, stop, step) serta print ketika angka tersebut ganjil, akan mengeluarkan teks "ini angka ganjil".

#####6. Anda memiliki data users = [
  {"id": 101, "role": "admin"},
  
  {"id": 102, "role": "user"},
  
  {"id": 101, "role": "admin"}
  
  ].

Ekstrak semua nilai id ke dalam sebuah list baru.

Ubah list id tersebut menjadi tipe data Set.

Cetak set tersebut dan jelaskan mengapa jumlah datanya berubah.

```python
users = [
{"id": 101, "role": "admin"},
{"id": 102, "role": "user"},
{"id": 101, "role": "admin"}
]

list_id = []

for biji in users:
    saringan_id = biji["id"]
    list_id.append(saringan_id)

set_id = set(list_id)

print("awal :", list_id)
print("ahir :", set_id)
```

items = [
  
  {"nama": "Meja", "stok": 5},

  {"nama": "Lampu", "stok": 12},
  
  {"nama" : "Monitor", "stok" : 2}

Gunakan for loop untuk mengiterasi data tersebut. Di dalam loop, gunakan if untuk mencetak kalimat "Nama Barang: [nama] perlu restock" HANYA jika stoknya kurang dari 4

```python
items = [

{"nama": "Meja", "stok": 5},

{"nama": "Kursi", "stok": 0},

{"nama": "Lampu", "stok": 12},

{"nama" : "Monitor", "stok" : 2}

]

for barang in items:
    if barang["stok"] < 4:
        print("namax :", barang["nama"], "restok gih")
```
